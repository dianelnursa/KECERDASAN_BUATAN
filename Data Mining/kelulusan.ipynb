```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
from sklearn import svm
import seaborn as sns; sns.set(font_scale=1.2)
import pickle
```

```python
datamhs = pd.read_csv('/content/drive/MyDrive/Classroom/Data_IPK_PIF_20.csv')
datamhs = datamhs.drop(['NIM','JENIS KELAMIN'], axis=1)
datamhs
```

```python
km=KMeans(n_clusters=2)
y_predicted=km.fit_predict(datamhs[['IPK']])
y_predicted
```

```python
datamhs['kelulusan']=y_predicted
```

```python
datamhs.to_csv('hasilkluster.csv')
```
